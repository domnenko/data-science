# Basics of Data Loading and Showing Summary Stats

```python
# libraries
import pandas as pd
import numpy as np
```

```python
# load the data set with bitcoin price, market cap, and trade volume
df = pd.read_csv(r'Inputs\btc-usd-max.csv')

# resolve dates
df['snapped_at'] = pd.to_datetime(
    df['snapped_at'],
    format='%Y-%m-%d %H:%M:%S %Z'
).dt.tz_convert(
    None
)
```

```python
# rename columns and create an index
df = df.rename(
    columns={'snapped_at': 'Date', 'price': 'Price', 'market_cap': 'Market Cap',
             'total_volume': 'Volume'}
).set_index(
    ['Date']
).copy()
```

```python
# show bitcoin price chart
plt = df[['Price']].plot()
```

```python
# show the market cap time series
plt = df[['Market Cap']].plot.area()
```

```python
# show the trade volume time series
plt = df[['Volume']].plot()
```

```python
# display basic summary stats
df.describe()
```
